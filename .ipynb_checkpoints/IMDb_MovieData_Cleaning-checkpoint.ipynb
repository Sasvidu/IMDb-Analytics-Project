```python
import pandas as pd
import numpy as np
```

```python
#Import the raw data csv
df = pd.read_csv('MostPopularIMDbMovies.csv')
```

```python
df
```

```python
#Convert runtime string value to an integer in minutes
def convertToMinutes(runtime):
  hours = 0
  minutes = 0

  # Split the string by whitespace
  parts = runtime.split()

  for part in parts:
    # Extract the number and unit (h or m)
    value, unit = int(part[:-1]), part[-1]

    if unit == 'h':
      hours = value
    elif unit == 'm':
      minutes = value
    else:
      raise ValueError(f"Invalid unit: {unit}")

  # Convert hours to minutes and add to total minutes
  minutesValue = hours * 60 + minutes

  return minutesValue
```

```python
#Remove upcoming movies
filtered_df = df[df['Year'].notna()]
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
#Remove plot synopsis column
filtered_df = filtered_df.drop('Plot_Synopsis', axis=1)
filtered_df
```

```python
#Create a new dataframe to store cleaned data
headings = ['Title', 'Year', 'Genre_1', 'Genre_2', 'Genre_3', 'Age_Rating', 'Runtime', 'Director', 'Star_1', 'Star_2', 'Star_3', 'IMDb_Rating', 'Metascore']
df = pd.DataFrame(columns = headings)

for index, movie in filtered_df.iterrows():
    title = movie['Title']
    
    year = movie['Year']

    #Split genre into 3 columns, so that data is atomic
    genres = movie['Genre']
    genres = genres.split('/')
    numberOfGenres = len(genres)
    genre1 = ""
    genre2 = ""
    genre3 = ""
    if numberOfGenres == 3:
        genre1 = genres[0]
        genre2 = genres[1]
        genre3 = genres[2]
    elif numberOfGenres == 2:
        genre1 = genres[0]
        genre2 = genres[1]
    elif numberOfGenres == 1:
        genre1 = genres[0]

    ageRating = movie['Age_Rating']

    #Convert Runtime into pure minutes, so it is quantitative
    runtime = movie['Runtime']
    runtime = convertToMinutes(runtime)

    director = movie['Director']

    #Split movie stars into 3 columns, so it is atomic
    stars = movie['Stars']
    stars = stars.split('/')
    numberOfStars = len(stars)
    star1 = ""
    star2 = ""
    star3 = ""
    if numberOfStars == 3:
        star1 = stars[0]
        star2 = stars[1]
        star3 = stars[2]
    elif numberOfStars == 2:
        star1 = stars[0]
        star2 = stars[1]
    elif numberOfStars == 1:
        star1 = stars[0]

    imdbRating = movie['IMDb_Rating']

    metascore = movie['Metascore']

    rowData = [title, year, genre1, genre2, genre3, ageRating, runtime, director, star1, star2, star3, imdbRating, metascore]

    length = len(df)
    df.loc[length] = rowData
```

```python
#Export the dataframe to a new csv
df.to_csv('CleanedMovies.csv', index=False, na_rep='')
```
