```python
#Scatterplot of IMDb Rating vs Metascore - show correlation of two ratings
imdb_rating = df['IMDb_Rating']
metascore = df['Metascore']

plt.figure(figsize=(8, 6))
sns.scatterplot(x=imdb_rating, y=metascore)

plt.xlabel('IMDb Rating')
plt.ylabel('Metascore')
plt.title('Scatter Plot - IMDb Rating vs Metascore')

plt.grid(True)

plt.show()
```

```python
#Boxplot showing rating distrubution according to the IMDb and Metascore scales

comparison_df = pd.DataFrame(columns = ['IMDb_Rating', 'Metascore'])

#Convert ratings to percentages
comparison_df['IMDb_Rating_Percent'] = df['IMDb_Rating'] * 10
comparison_df['Metascore_Percent'] = df['Metascore']

plt.figure(figsize=(18, 9))

ax = plt.subplot(1, 2, 1) 

sns.boxplot(x="variable", y="value", showmeans=True, data=comparison_df.melt(value_vars=['IMDb_Rating_Percent', 'Metascore_Percent']), ax=ax)

ax.set_xlabel('Rating Source')
ax.set_ylabel('Rating (%)')
ax.set_title('Distribution of Ratings (%)')

plt.xticks(rotation=45)  # Adjust rotation as needed

plt.tight_layout()

plt.show()
```

```python
#Perform ANOVA to compare ratings from IMDb vs Metascore
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('IMDb_Rating ~ Metascore', data=df).fit()
anova_results = anova_lm(model)
print(anova_results)
```

```python
#Most popular genre based on frequency of occurence
all_genres = pd.concat([df['Genre_1'], df['Genre_2'], df['Genre_3']], ignore_index=True)
genre_counts = all_genres.value_counts()
most_popular_genre = genre_counts.idxmax()
print(f"The most popular genre is: {most_popular_genre} (appearing {genre_counts.max()} times)")
```

```python
#Pie Chart of Genre Distribution
all_genres = pd.concat([df['Genre_1'], df['Genre_2'], df['Genre_3']], ignore_index=True)
genre_counts = all_genres.dropna().value_counts()
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.show()
```

```python
#Most popular director by frequency of movies directed
director_counts = df['Director'].value_counts()
most_popular_director = director_counts.idxmax()
popular_director_movies = df[df['Director'] == most_popular_director]['Title']
print(f"Most popular director is {most_popular_director}, who has directed ({director_counts.max()} movies):")
for movie in popular_director_movies:
    print("\t", movie)
```

```python
#Director with the highest IMDb Rated movies on average
director_groups = df.groupby('Director')
avg_ratings_by_director = director_groups['IMDb_Rating'].mean()
best_director = avg_ratings_by_director.idxmax()
highest_avg_rating = avg_ratings_by_director.max()
print(f"The director with the highest average IMDb rating is: {best_director} ({highest_avg_rating:.2f})")
```

```python
#Most Popular star acoording to frequency of movies starred in
all_stars = pd.concat([df['Star_1'], df['Star_2'], df['Star_3']], ignore_index=True).dropna()
star_counts = all_stars.value_counts()
most_popular_star = star_counts.idxmax()
max_movies = star_counts.max()
popular_star_movies = df[df['Star_1'].isin([most_popular_star]) | df['Star_2'].isin([most_popular_star]) | df['Star_3'].isin([most_popular_star])]

print(f"The most popular star is: {most_popular_star} (appearing in {max_movies} movies):")
for movie in set(popular_star_movies['Title']):
    print("\t", movie)
```

```python
#Predictive Analysis - Determine which genres are most likely to lead to high IMDb Ratings
genres = df[['Genre_1', 'Genre_2', 'Genre_3']].stack().unique()

genre_ratings = {}

for genre in genres:
    genre_df = df[(df['Genre_1'] == genre) | (df['Genre_2'] == genre) | (df['Genre_3'] == genre)]
    mean_rating = genre_df['IMDb_Rating'].mean()
    genre_ratings[genre] = mean_rating

top_genres = sorted(genre_ratings.items(), key=lambda x: x[1], reverse=True)[:3]
for genre, mean_rating in top_genres:
    print(f"{genre}: Mean IMDb Rating - {mean_rating:.2f}")
```

```python
#Predictive Analysis - Determine which genres are most likely to lead to low IMDb Ratings
genres = df[['Genre_1', 'Genre_2', 'Genre_3']].stack().unique()

genre_ratings = {}

for genre in genres:
    genre_df = df[(df['Genre_1'] == genre) | (df['Genre_2'] == genre) | (df['Genre_3'] == genre)]
    mean_rating = genre_df['IMDb_Rating'].mean()
    genre_ratings[genre] = mean_rating

worst_genres = sorted(genre_ratings.items(), key=lambda x: x[1])[:3]
for genre, mean_rating in worst_genres:
    print(f"{genre}: Mean IMDb Rating - {mean_rating:.2f}")
```
